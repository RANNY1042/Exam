```python
import pandas as pd
data={'before':[7,3,4,5,2,1,6,6,5,4],
      'after':[8,4,5,6,2,3,6,8,6,5]}
data=pd.DataFrame(data)
data
```

```python
stats.ttest_rel(data['after'],data['before'],alternative='greater')
```
